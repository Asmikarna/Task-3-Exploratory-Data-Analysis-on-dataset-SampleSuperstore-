```
col=['Postal Code']
df_new=df.drop(columns=col,axis=1)
```

```
df_new['State'].value_counts()
```

```
plt.figure(figsize=(15,15))
sns.countplot(x=df_new['State'])
plt.xticks(rotation=90)
plt.show()
```

```
df['Sub-Category'].value_counts()
plt.figure(figsize=(15,15))
sns.countplot(x=df_new['Sub-Category'])
plt.xticks(rotation=90)
plt.show()
```

```
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```
sns.heatmap(df.cov(),annot=True)
plt.show()
```

```
plt.figure(figsize=(20,20))
sns.barplot(x=df_new['Sub-Category'],y=df_new['Profit'])
plt.show()
```

```
figsize=(15,10)
sns.pairplot(df_new,hue='Sub-Category')
plt.show()
```

```
fig, ax=plt.subplots(figsize=(10,6))
ax.scatter(df_new["Sales"], df_new["Profit"])
ax.set_xlabel('Sales')
ax.set_ylabel('Profit')
plt.show()
```
